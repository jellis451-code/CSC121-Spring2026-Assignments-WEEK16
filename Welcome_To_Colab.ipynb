```
# PART 3: COMPLETION

expenses = []

while True:
    print("\n--- Expense Tracker ---")
    print("1. Add New Expense")
    print("2. Get Totals")
    print("3. Exit")

    selection = input("Selection: ")

    if selection == "1":
        category = input("Enter Category: ")
        vendor = input("Enter Vendor: ")

        while True:
            try:
                amount = float(input("Enter Amount: "))
                break
            except ValueError:
                print("Invalid amount. Enter a number.")

        date = input("Enter Date: ")

        expense = {
            "category": category,
            "vendor": vendor,
            "amount": amount,
            "date": date
        }

        expenses.append(expense)
        print("Expense added!")

    elif selection == "2":
        print("\nA: All Expenses")
        print("C: By Category")
        print("V: By Vendor")

        option = input("Choice: ").upper()

        if option == "A":
            total = sum(e["amount"] for e in expenses)
            print(f"Total: ${total:.2f}")

        elif option == "C":
            cat = input("Enter Category: ")
            total = sum(e["amount"] for e in expenses if e["category"].lower() == cat.lower())
            print(f"{cat} Total: ${total:.2f}")

        elif option == "V":
            ven = input("Enter Vendor: ")
            total = sum(e["amount"] for e in expenses if e["vendor"].lower() == ven.lower())
            print(f"{ven} Total: ${total:.2f}")

    elif selection == "3":
        print("Goodbye!")
        break
```
