```
# PART 2: CORE LOGIC

expenses = []

while True:
    print("\n1. Add Expense")
    print("2. Exit")

    choice = input("Choice: ")

    if choice == "1":
        category = input("Category: ")
        vendor = input("Vendor: ")
        amount = float(input("Amount: "))
        date = input("Date: ")

        expense = {
            "category": category,
            "vendor": vendor,
            "amount": amount,
            "date": date
        }

        expenses.append(expense)

        print("Current Expenses List:")
        print(expenses)

    elif choice == "2":
        break
```
